```
def type_count(df): 
    df1 = df.groupBy("type").count().orderBy('count', ascending=False)
    return df1

counts_2019 = type_count(df_2019)
counts_2020 = type_count(df_2020)
counts_2021 = type_count(df_2021)
```

```
x_axis = counts_2019.select('type').rdd.flatMap(lambda x: x).collect()
y1 = counts_2019.select('count').rdd.flatMap(lambda x: x).collect()
y2 = counts_2020.select('count').rdd.flatMap(lambda x: x).collect()
y3 = counts_2021.select('count').rdd.flatMap(lambda x: x).collect()
```

```
from bokeh.plotting import figure
from bokeh.embed import components, file_html
from bokeh.resources import CDN

colors = ["#c9d9d3", "#718dbf", "#FF5733" ]

data = {"names" : x_axis, 
        '2019': y1, 
        '2020': y2,
        '2021': y3}

versions = ['2019', '2020', '2021']


p = figure(x_range=x_axis,height=250, title="Types of Studies", toolbar_location="below")

p.vbar_stack(versions, x='names', width=1, color=colors, source=data,
             legend_label=versions)
p.left[0].formatter.use_scientific = False

html = file_html(p, CDN, "my plot1")
displayHTML(html)
```

```
from pyspark.sql.functions import * 

def count_conditions(df):
    df1 = df.select("conditions").withColumn('split', split(df["conditions"],","))
    df2= df1.withColumn('exploded1', explode(df1.split)).select('exploded1').groupBy('exploded1').count().orderBy('count', ascending=False)
    return df2
```

```
def completed(df): 
    new = df.select(['Completion', 'Status']).where(df['Status'] == 'Completed')
    return new 

q6_2019 = completed(df_2019)
q6_2020 = completed(df_2020)
q6_2021 = completed(df_2021)


display(q6_2019)
```

```
def split_months(df, year): 
    df1 = df.withColumn('Month', split(df['Completion'], ' ').getItem(0)).withColumn('year', split(df['Completion'], ' ').getItem(1))
    df2 = df1.where(df1.year == year).select('Month').groupBy('Month').count() 
    return df2 
q6_2019_1 = split_months(q6_2019, "2019")
q6_2020_1 = split_months(q6_2020, "2020")
q6_2021_1 = split_months(q6_2021, "2021")

q6_2019_1.show()
```
